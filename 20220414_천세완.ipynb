1-1. 빅데이터의 3V에 대해 설명하시오.
3V
거대한 규모(volume) : 방대한 양의 데이터(테라, 페타바이트 이상)
빠른 속도(velocity) : 실시간 생성, 빠른 속도로 데이터를 처리/분석
높은 다양성(variety) : 정형(DBMS 등) 및 비정형(영상, 음성, 텍스트 SNS 등)

1-2. 정형데이터와 비정형데이터에 대해 설명하시오.
정형 데이터
- 일정한 형식이나 규칙에 맞게 저장된 데이터
비정형 데이터
- 정해진 규칙이 없어서 값의 의미를 쉽게 파악하기 힘든 데이터들

1-3. 빅데이터의 처리 과정에 대해 설명하시오.
데이터 소스 -> 수집 -> 저장 -> 처리 -> 분석 -> 표현

1-4. 빅데이터의 활용 사례를 2가지 이상 설명하시오.
당근마켓
- 게시한 사진의 상품 카테고리 자동 분류, 맞춤형 콘텐츠 추천
넷플릭스
